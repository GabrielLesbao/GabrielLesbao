**Carregar as bibliotecas utilizadas**


```python
#dataframe
import pandas as pd

#web scraping
from bs4 import BeautifulSoup
import requests

#Ferramentas para processar linguagem natural
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords 

#data viz
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from wordcloud import WordCloud
plt.style.use('ggplot')
```

OBTER OS TEXTOS DA URL


```python
url = "https://biopython.org/wiki/Documentation"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
soup = soup.get_text()

soup[-1000:]
```

```python
tokens = [tok for tok in soup.split()]
print(tokens[:10])
print(len(tokens))
```

Frequency distribution
**negrito**

```python
Freq_dist_nltk=nltk.FreqDist(tokens)
# for a,b in Freq_dist_nltk.items():
#     print(str(a)+':'+str(b))
```

```python
plt.figure(figsize=(20,5))
ax = Freq_dist_nltk.plot(50, cumulative=False)
# below is the plot for the frequency distributions
```

```python
stop_words = set(stopwords.words('english')) 

result = []
for word in tokens: 
    word = word.lower()
    if word not in stop_words: 
        result.append(word) 
        
clean_dist_nltk=nltk.FreqDist(result)
```

```python
plt.figure(figsize=(20,5))
clean_dist_nltk.plot(25, cumulative=False,color='blue')
```

Convert into dataframe

```python
df = pd.DataFrame.from_dict(clean_dist_nltk, orient='index')
df.columns = ['Frequency']
df.index.name = 'Term'
df = df.sort_values(by=['Frequency'],ascending=False)
df = df.reset_index()
```

```python
df.head()
```

 boxplot

```python
plt.figure(figsize=(20,5))
df1 = df.iloc[0:10]
sns.barplot(x = df1['Term'], 
            y = df1['Frequency'],
            palette="BuGn_r")
```

wordcloud

```python
data = dict(zip(df['Term'].tolist(), df['Frequency'].tolist()))

data = df.set_index('Term').to_dict()['Frequency']

# print(data)
plt.figure(figsize = (20,15))
wc = WordCloud(width=800, 
               height=500, 
               max_words=200,
               colormap="BuGn_r",
               background_color="black").generate_from_frequencies(data)


plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
